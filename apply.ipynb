```python
apply on df will take one column at a time  eg. /// x.max() - x.min()  OR ///all element x*2
apply on df[col] will take one row at a time XXXXeg. x.max() - x.min() ////eg. x*2
applymap work on all element XXXXeg. x.max() - x.min()  ///// all element x*2
map same as apply for row
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(4, 3), columns=list('bde'))
df
```

```python
f = lambda x: x.max() - x.min()
```

```python
df.apply(f)
```

```python
df['b'].apply(f)
```

```python
df.applymap(lambda x: 10*x)
```

```python
df.applymap(f)
```

```python
df['b'].applymap(lambda x: 10*x)
```

```python
df.apply(lambda x: 10*x)
```

```python
df.map(lambda x: 10*x)
```

```python
df['b'].map(lambda x: 10*x)
```
